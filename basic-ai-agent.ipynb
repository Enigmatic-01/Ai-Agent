```python
import json
import requests
from langchain_openai import ChatOpenAI
from langchain_classic.agents import create_react_agent,AgentExecutor
from langchain_core.tools import tool
from langchain_community.tools import DuckDuckGoSearchRun
from langchain_classic import hub
```

LLM

```python
llm = ChatOpenAI(
    api_key=OPENAI_KEY
)
```

Tools

```python
search_tool = DuckDuckGoSearchRun()


@tool
def get_weather_tool(city:str) -> float:
    "get_weather_tool take input the name of place and return the weather information"
    url = f"https://api.openweathermap.org/data/2.5/weather?q={city}&appid={WEATHER_API_KEY}"
    res = requests.get(url)
    return res.json()
```

ReAct Prompt


```python
react_prompt = hub.pull("hwchase17/react")
```

React Agent Object

```python
agent = create_react_agent(
    llm=llm,
    prompt=react_prompt,
    tools=[search_tool,get_weather_tool]
)
```

Agent Executor Object

```python
agent_executor = AgentExecutor(
    agent= agent,
    tools=[search_tool,get_weather_tool],
    verbose=True
    
)
```

Query from User

```python
query  = "Using real-time data, tell me which city in India currently has the highest air pollution levels (PM2.5 or AQI) and provide today’s weather for that city, including temperature, humidity, and general conditions."
```

Output

```python
agent_output=agent_executor.invoke({"input":query})
```
